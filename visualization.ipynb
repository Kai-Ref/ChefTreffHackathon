```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df_actual = pd.read_csv('datasets/first_tier_prices_actual.csv')
df_forecast = pd.read_csv('datasets/first_tier_prices_forecast.csv')

# Convert the 'datetime_utc' column to datetime type
df_actual['datetime_utc'] = pd.to_datetime(df_actual['datetime_utc'])
df_forecast['datetime_utc'] = pd.to_datetime(df_forecast['datetime_utc'])

# Plot the time series
plt.plot(df_actual['datetime_utc'], df_actual['energy_costs_ct_per_kWh'])
plt.plot(df_forecast['datetime_utc'], df_forecast['forecasted_energy_costs_ct_per_kWh'], alpha=0.5)
plt.xlabel('Date')
plt.ylabel('Energy Costs (ct/kWh)')
plt.title('Energy Costs Time Series')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.plot(df_actual['datetime_utc'], df_actual['energy_costs_ct_per_kWh'] - df_forecast['forecasted_energy_costs_ct_per_kWh'])
plt.xlabel('Date')
plt.ylabel('Diff Energy Costs (ct/kWh)')
plt.title('Energy Costs Time Series')
plt.xticks(rotation=45)
plt.show()

(df_actual['energy_costs_ct_per_kWh'] - df_forecast['forecasted_energy_costs_ct_per_kWh']).abs().mean()
```
